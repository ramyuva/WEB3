```python
driver.get(url)
```

```python
date=[]
name=[]
industry=[]
sub_vertical=[]
city=[]
inname=[]
intype=[]
amount=[]
```

```python
# 35:64 is taken in every column because from july to september there between 35:64 only
titles_tag1=driver.find_elements_by_xpath("//td[@class='column-2']")
for i in titles_tag1:
    date.append(i.text)
date[35:64]
```

```python
titles_tag2=driver.find_elements_by_xpath("//td[@class='column-3']")
for i in titles_tag2:
    name.append(i.text)
name[35:64]
```

```python
titles_tag3=driver.find_elements_by_xpath("//td[@class='column-4']")
for i in titles_tag3:
    industry.append(i.text)
industry[35:64]
```

```python
titles_tag4=driver.find_elements_by_xpath("//td[@class='column-5']")
for i in titles_tag4:
    sub_vertical.append(i.text)
sub_vertical[35:64]
```

```python
titles_tag5=driver.find_elements_by_xpath("//td[@class='column-6']")
for i in titles_tag5:
    city.append(i.text)
city[35:64]
```

```python
titles_tag6=driver.find_elements_by_xpath("//td[@class='column-7']")
for i in titles_tag6:
    inname.append(i.text)
inname[35:64]
```

```python
titles_tag7=driver.find_elements_by_xpath("//td[@class='column-8']")
for i in titles_tag7:
    intype.append(i.text)
intype[35:64]
```

```python
titles_tag8=driver.find_elements_by_xpath("//td[@class='column-9']")
for i in titles_tag8:
    amount.append(i.text)
amount[35:64]
```

```python
import pandas as pd
trak=pd.DataFrame()
trak['Date']=date[35:64]
trak['STRATUP_NAME']=name[35:64]
trak['INDUSTRY']=industry[35:64]
trak['SUB_VERTICAL']=sub_vertical[35:64]
trak['CITY']=city[35:64]
trak['INVESTOR_NAME']=inname[35:64]
trak['INVESTMENT_TYPE']=intype[35:64]
trak['AMOUNT']=amount[35:64]
```
