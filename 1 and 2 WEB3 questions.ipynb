```python
# Function Definition
def amazon_mob(url):
    driver=webdriver.Chrome('chromedriver.exe')
    start_page=0
    end_page=2
    urls = []
    brand_name=[]
    name=[]
    price=[]
    product_url=[]
    rating=[]
    no_of_ratings=[]
    exchange=[]
    delivery=[]
    availability=[]
    Other_Details=[]
    
    #loop to fetch urls of each mobile till page 5
    for page in range(start_page,end_page+1):
        driver.get(url)
        soup= BeautifulSoup(driver.page_source, 'html.parser')
        prod_urls = soup.find_all('a', attrs ={'class':'a-link-normal a-text-normal'})
        for prod in prod_urls:
            urls.append('https://www.amazon.in'+prod.get('href'))
    
    #loop to scrap required details from each mobile page
    for url in urls:
        driver.get(url)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        brand = soup.find('h1',attrs={'id':'title'})
        if brand is not None:
            brand_name.append(brand.find('span').text[0:8].replace('\n',''))
        else:
            brand_name.append('-')
        n = soup.find('h1',attrs={'id':'title'})
        if n is not None:
            name.append(n.find('span').text.replace('\n',''))
        else:
            name.append('-')
        rat = soup.find('div', attrs = {'id':'averageCustomerReviews'})
        if rat is not None:
            rating.append(rat.find('i').find('span').text)
        else:
            rating.append('-')
        rate = soup.find('span', attrs = {'id':'acrCustomerReviewText'})
        if rate is not None:
            no_of_ratings.append(rate.text)
        else:
            no_of_ratings.append('-')    
        p = soup.find('span', attrs = {'class':'a-size-medium a-color-price priceBlockDealPriceString'})
        if p is not None:
            price.append(p.text[2:])
        else:
            p = soup.find('div', attrs = {'id':'price'})
            if p is not None:
                price.append(p.find('span').text[2:])
            else:
                price.append('-')
        url = soup.find('div', attrs = {'class':'imgTagWrapper'})
        if url is not None:
            product_url.append(url.find('img').get('src'))
        else:
            product_url.append('-')
        exc = soup.find('a',attrs={'class':'a-size-small a-link-normal a-text-normal'})
        if exc is not None:
            exchange.append(exc.text.replace('\n',''))
        else:
            exchange.append('-')   
            name.append('-')
        deli = soup.find('div',attrs={'class':'a-section a-spacing-mini'})
        if deli is not None:
            delivery.append(deli.text)
        else:
            delivery.append('-')   
        ave = soup.find('div', attrs = {'class':'a-section a-spacing-mini'})
        if ave is not None:
            availability.append(ave.text.replace('\n',''))
        else:
            availability.append('-')   
        oth = soup.find('span', attrs = {'class':'selection'})
        if oth is not None:
            Other_Details.append(oth.text.replace('\n',''))
        else:
            Other_Details.append('-')                
    mob_df = df=pd.DataFrame({'Brand_name':brand_name[0:50],
                              'Name':name[0:50],
                              'Price':price[0:50],
                              'Rating':rating[0:50],
                              'Product_URL':product_url[0:50],
                              'No_of_Ratings':no_of_ratings[0:50],
                              'Exchange':exchange[0:50],
                              'Delivery':delivery[0:50],
                              'Availability':availability[0:50],
                              'Other_Details':Other_Details[0:50]})
    print(mob_df)
    mob_df.to_csv('Amazon Mobiles.csv', index = False)
    
    
# Calling Function
amazon_mob('https://www.amazon.in/s?k=mobile+phone+under+20000&ref=nb_sb_noss_1')
```
